```python
# this function iterates over a folder with txt files and copies and pastes all the files whose names are written in a list with file names to store them in a new folder

import os
import shutil

def copy_files(source_dir, destination_dir, file_list):
    # Create the destination directory if it doesn't exist
    if not os.path.exists(destination_dir):
        os.makedirs(destination_dir)

    # Iterate through the list of filenames
    for filename in file_list:
        source_file_path = os.path.join(source_dir, filename)
        destination_file_path = os.path.join(destination_dir, filename)
        
        # Check if the file exists in the source directory
        if os.path.exists(source_file_path):
            # Copy the file to the destination directory
            shutil.copy2(source_file_path, destination_file_path)
            print(f"Copied '{filename}' to '{destination_dir}'")
        else:
            print(f"File '{filename}' not found in '{source_dir}'")

# Example usage
source_directory = "/Users/sguhr/Desktop/Diss_Korpus/20240313_Diss_Corpus"
destination_directory = "/Users/sguhr/Desktop/Diss_Korpus/chosen_files"
file_names_list = ["Boelte_Amely_Die_Candidaten-Braut.txt", "Baumbach_Rudolf_Aus_der_Jugendzeit_Das_Habichtsfraeulein.txt", "Gerstaecker_Friedrich_Die_Gemsjagd.txt", "Berend_Alice_Die_Braeutigame_der_Babette.txt", "von_Sternberg_Alexander_Ein_literarischer_Abend_des_alten_Dresden.txt", "Berthold_Theodor_Lustige_Gymnasialgeschichten_Der_Hauptgewinn.txt", "Storm_Theodor_Im_Schloss.txt", "Willkomm_Ernst_Die_Echte.txt", "Ebers_Georg_Die_Geschichte_meines_Lebens.txt", "Dohm_Hedwig_Werde.txt", "David_Jakob_Julius_Die_Hanna.txt", "Engel_Georg_Der_verbotene_Rausch_Onkel_Poekel.txt", "Ernst_Otto_Ein_frohes_Farbenspiel.txt", "Flex_Walter_Wolf_Eschenlohr.txt", "Fontane_Theodor_Cecile.txt", "Frapan_Ilse_Bittersuess_Frauenliebe.txt", "Fritze_Ernst_Lord_Felix.txt", "Grasberger_Hans_Maria-Buch.txt", "Gottschall_Rudolf_Die_zehnte_Sprache.txt", "Groller_Balduin_Detektiv_Dagobert_Empfang_beim_Ministerpraesidenten.txt", "Hart_Heinrich_Zwei_Skizzenblaetter_Vincenz.txt", "Hawel_Rudolf_Dr_Thorns_Lebensabend.txt", "Heigel_Carl_August_Er_kommt_nicht.txt", "Horn_Georg_Irrende_Sterne.txt", "Janitschek_Maria_Hysterie.txt", "Koenig_Alma_Johanna_Schibes.txt", "Koch_Henny_Klein-Grosschen.txt", "Lingg_Hermann_Nicht_zu_hoch.txt", "Loens_Hermann_Tiergeschichten_Der_Moerder.txt", "Mann_Heinrich_Professor_Unrat.txt", "Marlitt_Eugenie_Im_Hause_des_Commerzienrathes.txt", "Meisel-Hess_Grete_Die_Intellektuellen.txt", "Musil_Robert_Die_Verwirrungen_des_Zoeglings.txt", "Proelss_Johannes_Lili.txt", "Raabe_Wilhelm_Drei_Federn_Auch_Mathildes_Tage.txt", "Rie_Therese_Die_Liebende_Die_Liebende.txt", "Rilke_Rainer_Maria_Die_Geschwister.txt", "Ringelnatz_Joachim_Fahrensleute.txt", "Schrader_August_Klementine.txt", "Schnitzler_Arthur_Frau_Beate.txt", "Spyri_Johanna_Heidis_Lehr-_und_Wanderjahre.txt", "Storch_Ludwig_Das_Schachtgespenst.txt", "Storm_Theodor_Ein_stiller_Musikant.txt", "Sudermann_Clara_Schatten.txt", "Temme_Jodocus_Auf_der_Eisenbahn.txt", "Thoma_Ludwig_Franz_und_Cora.txt", "Villinger_Hermine_Aus_unserer_Zeit_Die_Bas.txt", "Villinger_Hermine_Aus_unserer_Zeit_Die_Bas.txt", "von_Eschstruth_Nataly_Der_Stern_des_Gluecks.txt", "von_Francois_Louise_Judith_die_Kluswirtin.txt", "von_Perfall_Anton_Das_verlorene_Paradies.txt", "von_Peternell_Anna_Die_Frau_Majorin.txt", "von_Wolzogen_Ernst_Vom_Peperl_und von andern_Raritaeten_Der_Bluethner_Fluegel.txt", "Weiss_Ernst_Franziska.txt", "Wothe_Anny_Durch_Graue_Gassen.txt", "zu_Reventlow_Franziska_Herrn_Dames_Aufzeichnungen.txt", "Zweig_Stefan_Die_Liebe_der_Erika.txt", "Linz_Amelie_Kein_Herz.txt"] # List of file names to copy

copy_files(source_directory, destination_directory, file_names_list)
```

Take the chosen files and automatically transform them into xml files.

```python
# with this function every txt file in a given folder are copied and converted into xml files stored in a new folder following the indicated xml structure.

import os

def remove_ds_store_file(source_dir):
    # Remove the .DS_Store file if it exists
    ds_store_path = os.path.join(source_dir, ".DS_Store")
    if os.path.exists(ds_store_path):
        os.remove(ds_store_path)
        print(".DS_Store file removed")


def transform_to_xml(source_dir, destination_dir):
    # Iterate through each file in the source directory
    for filename in os.listdir(source_dir):
        source_file_path = os.path.join(source_dir, filename)
        
        # Check if the file is a regular file
        if os.path.isfile(source_file_path):
            # Read the content of the file
            with open(source_file_path, 'r') as file:
                content = file.read()
            
            # Construct the XML content
            xml_content = f"""<?xml version="1.0" encoding="UTF-8"?>
<TEI xmlns="http://www.tei-c.org/ns/1.0" xmlns:xs="http://www.w3.org/2001/XMLSchema">
    <teiHeader>
        <fileDesc>
            <titleStmt>
                <title>{filename}</title>
            </titleStmt>
        </fileDesc>
    </teiHeader>
    <text>
        <body>
{content}
        </body>
    </text>
</TEI>
"""

            # Write the XML content to a new file in the destination directory
            xml_file_path = os.path.join(destination_dir, filename.replace('.txt', '.xml'))
            with open(xml_file_path, 'w') as xml_file:
                xml_file.write(xml_content)
            
            print(f"File '{filename}' transformed to XML and saved as '{xml_file_path}'")

# Example usage
source_directory = "/Users/sguhr/Desktop/Diss_Korpus/20240401_Diss_Corpus"
destination_directory = "/Users/sguhr/Desktop/Diss_Korpus/20240404_Diss_Corpus_xml"

# Optionally remove the .DS_Store file
remove_ds_store_file(source_directory)

# Transform files to XML
transform_to_xml(source_directory, destination_directory)
```
